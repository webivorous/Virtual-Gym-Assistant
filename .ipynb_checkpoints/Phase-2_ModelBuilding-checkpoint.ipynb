```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

import sklearn
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
from sklearn.metrics import accuracy_score, recall_score, precision_score ,f1_score
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, confusion_matrix

from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB, GaussianNB
from sklearn.neural_network import MLPClassifier

import pickle
from joblib import dump, load

import cv2
import mediapipe as mp
from sklearn.metrics import classification_report
import os
```

# Feature extraction: computing angle


```python
def get_angle(frame, keypoint_1, keypoint_2, keypoint_3):
    a = np.array([frame['x'+str(keypoint_1)], frame['y'+str(keypoint_1)]]) # First
    b = np.array([frame['x'+str(keypoint_2)], frame['y'+str(keypoint_2)]]) # Mid
    c = np.array([frame['x'+str(keypoint_3)], frame['y'+str(keypoint_3)]]) # End
    
    ba = a - b
    bc = c - b

    cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc))
    angle = math.degrees(np.arccos(cosine_angle))
    
    if angle > 180.0:
        angle = 360 - angle
        
    return angle
```

```python
def get_y_difference(frame, keypoint_1, keypoint_2):
    return abs(frame['y'+str(keypoint_1)] - frame['y'+str(keypoint_2)])
```

# --------------------------------------------------Squat----------------------------------------------------

```python
landmarks_df = pd.read_csv('squat-main.csv')
landmarks_df
```

```python

landmarks_df['hip_angle'] = landmarks_df.apply(lambda row: get_angle(row, 11, 23, 25), axis=1)
landmarks_df['knee_hip_difference'] = landmarks_df.apply(lambda row: get_y_difference(row, 23, 25), axis=1)
landmarks_df['knee_angle'] = landmarks_df.apply(lambda row: get_angle(row, 23, 25, 27), axis=1)
```

```python
landmarks_df
```

# Spliting into train and testing

```python
target=landmarks_df.columns[0]
features=['hip_angle','knee_hip_difference','knee_angle']

X=landmarks_df[features]
y=landmarks_df[target]
target_names=list(set(y))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# SVM

```python
svc_model = LinearSVC(random_state=1).fit(X_train, y_train)
y_pred = svc_model.predict(X_test)

svm_report=classification_report(y_test, y_pred,target_names=target_names)
print(svm_report)
```

```python
dump(svc_model, 'svmSquat.joblib')
```

# KNN

```python
knn_model = KNeighborsClassifier(25).fit(X_train, y_train)
y_pred = knn_model.predict(X_test)

knn_report=classification_report(y_test, y_pred,target_names=target_names)
print(knn_report)
```

```python
dump(knn_model, 'knnSquat.joblib') 
```

# Nave Bayes

```python
gnb_model = GaussianNB()
y_pred = gnb_model.fit(X_train, y_train).predict(X_test)

gnb_report=classification_report(y_test, y_pred,target_names=target_names)
print(gnb_report)
```

```python
dump(gnb_model, 'gnbSquat.joblib')
```

# Neural Network

```python
mlp_clf = MLPClassifier(activation='relu',
                        learning_rate='adaptive',
                        hidden_layer_sizes=(100),
                        learning_rate_init=0.01,
                        random_state=1)

y_pred = mlp_clf.fit(X_train, y_train).predict(X_test)

mlp_report=classification_report(y_test, y_pred,target_names=target_names)
print(mlp_report)
```

```python
dump(mlp_clf, 'mlp-squat.joblib')
```

# -----------------------------------------Overhead Press-----------------------------------------------

```python
landmarks_df = pd.read_csv('overhead-main.csv')
landmarks_df
```

```python
landmarks_df['bicep_angle'] = landmarks_df.apply(lambda row: get_angle(row, 11, 13, 15), axis=1)
landmarks_df['hip_angle'] = landmarks_df.apply(lambda row: get_angle(row, 11, 23, 25), axis=1)
landmarks_df['knee_angle'] = landmarks_df.apply(lambda row: get_angle(row, 23, 25, 27), axis=1)
landmarks_df['foot_angle'] = landmarks_df.apply(lambda row: get_angle(row, 25, 27, 31), axis=1)
landmarks_df['elbow_hip_angle'] = landmarks_df.apply(lambda row: get_angle(row, 13, 11, 23), axis=1)
```

```python
landmarks_df
```

```python
target=landmarks_df.columns[0]
features=['hip_angle','bicep_angle','knee_angle','foot_angle','elbow_hip_angle']

X=landmarks_df[features]
y=landmarks_df[target]
target_names=list(set(y))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# SVC

```python
svc_model = LinearSVC(random_state=1).fit(X_train, y_train)
y_pred = svc_model.predict(X_test)

svm_report=classification_report(y_test, y_pred,target_names=target_names)
print(svm_report)
```

```python
dump(svc_model, 'SVC-overhead.joblib')
```

# Nearest Neighbours

```python
error_rate = []
for i in range(1,35):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(1,35),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.show()
```

```python
k=5
knc_model = KNeighborsClassifier(k).fit(X_train, y_train)
y_pred = knc_model.predict(X_test)

knc_report=classification_report(y_test, y_pred,target_names=target_names)
print(knc_report)
```

```python
dump(knc_model, 'knc-overhead.joblib')
```

# Naive Bayes

```python
gnb_model = GaussianNB()
y_pred = gnb_model.fit(X_train, y_train).predict(X_test)

gnb_report=classification_report(y_test, y_pred,target_names=target_names)
print(gnb_report)
```

```python
mnb_model = GaussianNB()
y_pred = gnb_model.fit(X_train, y_train).predict(X_test)

mnb_report=classification_report(y_test, y_pred,target_names=target_names)
print(gnb_report)
```

```python
dump(mnb_model, 'mnb-overhead.joblib')
```

# Neural Network

```python
mlp_clf = MLPClassifier(activation='relu',
                        learning_rate='adaptive',
                        hidden_layer_sizes=(100),
                        learning_rate_init=0.01,
                        random_state=1)

y_pred = mlp_clf.fit(X_train, y_train).predict(X_test)

mlp_report=classification_report(y_test, y_pred,target_names=target_names)
print(mlp_report)
```
